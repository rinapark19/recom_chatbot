```python
from langchain.llms import OpenAI
llm1 = OpenAI(model_name="text-davinci-003", temperature=0.9)
llm1("초등학생이 읽을 만한 책 추천해 줘")
```

```python
llm2 = OpenAI(model_name="gpt-3.5-turbo", temperature=0.9)
llm2("초등학생이 읽을 만한 책 추천해 줘")
```

```python
def get_responses(prompt:str):
  data = {}
  llm1 = OpenAI(model_name="text-davinci-003", temperature=0.9)
  llm2 = OpenAI(model_name="gpt-3.5-turbo", temperature=0.9)
  data['prompt'] = prompt
  data['response1'] = llm1(prompt)
  data['response2'] = llm2(prompt)
  return data
```

```python
data = get_responses("초등학생이 읽을 만한 책 추천해 줘")
data
```

```python
def choose(number):
  if number == 1:
    data['chosen'] = data.pop('response1')
    data['rejected'] = data.pop('response2')

  elif number == 2:
    data['chosen'] = data.pop('response2')
    data['rejected'] = data.pop('response1')
```

```python
choose(2)
```
