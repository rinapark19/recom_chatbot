```python
import pandas as pd
```

```python
data = pd.read_csv("/content/drive/MyDrive/SCM_books/SCM_books.csv")
```

```python
data
```

**1, DataFrame Column 붙이기**

```python
columns = ['author', 'category', 'introduction', 'isbn', 'publish_date', 'publisher', 'title', 'toc']
data.columns = columns
```

```python
data
```

**1.1 column 이름 바꾸면서 날아간 데이터 추가**

```python
new_row = {'author': '이지관', 'category': '종교 > 불교 > 불교철학/사상', 'introduction': '이 책은 일반인들이 불교에 대해 이해할 수 있는 교양서이다.', 'isbn': '2000127000059',
          'publish_date': '1976-08-01', 'publisher': '가산문고', 'title': '남북전육부 율장비교연구(3판)', 'toc': '이 상품은 목차가 없습니다.'}
```

```python
data_new = data.append(pd.Series(), ignore_index=True)
data_new.loc[1:] = data_new.loc[:].shift()

data_new.loc[0] = new_row
```

```python
data_new
```

**2. 결측값 채우기**

```python
df_front = data_new[['author', 'category', 'introduction', 'isbn', 'publish_date', 'publisher', 'title']]
df_back = data_new['toc']
```

```python
df_front.fillna("없음", inplace=True)
```

```python
df_back.fillna("이 상품은 목차가 없습니다.", inplace=True)
df_back
```

```python
df = pd.concat([df_front, df_back], axis = 1)
```

```python
df
```

**3. ISBN, 제목 기준 중복값 제거**

```python
final_df = df.drop_duplicates(subset=['isbn', 'title'])
```

```python
final_df
```

**4. ISBN 형식 맞춰 주기**

```python
def cleaning(x):
  x = str(x)
  if '.' in x:
    return x[:x.index('.')]
  else:
    return x
```

```python
final_df['isbn'] = final_df['isbn'].apply(cleaning)
```

```python
final_df
```

```python
final_df.to_csv("book_info.csv", encoding="utf8")
```
