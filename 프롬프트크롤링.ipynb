```python
for url in URL_list:
  # url 긁어오기
  source_code_from_URL = urllib.request.urlopen(url)
  soup = BeautifulSoup(source_code_from_URL, "lxml", from_encoding="utf8")

  # 질문 부분만 리스트에 넣기
  for item in soup.find_all('div', class_="question_group"):
    prompt_list.append(item.text.strip())
```
